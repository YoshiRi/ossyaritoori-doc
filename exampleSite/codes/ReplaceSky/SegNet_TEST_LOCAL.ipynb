```python
import numpy as np
import keras
import cv2
import matplotlib.pyplot as plt
from importlib import reload
reload(keras.models)
```

```python
height = 360
width = 480
classes = 12
data_shape = 360*480

def predict(test):
    model = keras.models.load_model('seg.h5')
    probs = model.predict(test, batch_size=1)

    prob = probs[0].reshape((height, width, classes)).argmax(axis=2)
    return prob
```

```python
fuji = cv2.imread('fuji.jpg')
plt.imshow(fuji)
```

```python
from keras.layers import Input
from keras.layers.core import Activation, Flatten, Reshape
from keras.layers.convolutional import Convolution2D, Conv2D, MaxPooling2D, UpSampling2D
from keras.layers.normalization import BatchNormalization
from keras.models import Model
from keras.utils import np_utils


import h5py

def ExtractSky(img):
    hei,wid,_=img.shape
    #360 480
    M = np.array([480.0/wid, 0, 0, 0, 360.0/hei, 0]).reshape(2,3)
    dst = cv2.warpAffine(img,M,(480,360))
    plt.imshow(dst)
    predicted = predict(dst)
    predicted
    
ExtractSky(fuji)
```
